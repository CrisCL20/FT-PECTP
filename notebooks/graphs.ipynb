```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
sns.set_style("whitegrid")
colors = ["#DC1919", '#1F73FA']
sns.set_theme(context='paper', palette=sns.color_palette(colors))
```

```python
times=[10898, 22661]
cases = ["Caso 1", "Caso 2"]
df = pd.DataFrame({
    "case" : cases,
    "times" : times
})

ax = sns.barplot(df, x="case", y="times", estimator="sum")
plt.title("Tiempos de ejecución de cada caso de prueba")
plt.ylabel("Tiempo [s]")
plt.xlabel("")

for container in ax.containers:
    ax.bar_label(container)

plt.show()
```

```python
nsga_results_i4 = [
[1.100000e+01, 3.000000e+01],
[6.000000e+00, 4.100000e+01],
[1.700000e+01, 2.100000e+01],
[9.000000e+00, 3.200000e+01],
[1.000000e+01, 3.100000e+01],
[1.300000e+01, 2.900000e+01],
[1.500000e+01, 2.600000e+01],
[3.000000e+01, 1.000000e+01],
[2.300000e+01, 1.800000e+01],
[1.300000e+01, 2.900000e+01],
[1.400000e+01, 2.700000e+01],
[7.000000e+00, 3.700000e+01],
[1.600000e+01, 2.500000e+01],
[5.000000e+00, 4.600000e+01],
[2.300000e+01, 1.800000e+01],
[6.000000e+00, 4.100000e+01],
[2.500000e+01, 1.500000e+01],
]
nsga_results_i4 = np.array(nsga_results_i4)
```

```python
type_ampl = np.full((len(objectives_i4_ampl),1), fill_value="AMPL")
type_nsga = np.full((len(nsga_results_i4),1), fill_value="NSGA-II")
ampl_results_toy=np.concatenate((objectives_i4_ampl, type_ampl), axis=1)
nsga_results_toy=np.concatenate((nsga_results_i4, type_nsga), axis=1)
nsga_results_toy.shape
```

```python
toy_results = np.vstack((nsga_results_toy, ampl_results_toy))
```

```python
df = pd.DataFrame(toy_results, columns=["z1", "z2", "Algoritmo"])
df["z1"] = pd.to_numeric(df["z1"])
df["z2"] = pd.to_numeric(df["z2"])
ax = sns.scatterplot(df, x="z2", y="z1", hue="Algoritmo")
plt.ylabel(r"$z_1$")
plt.xlabel(r"$z_2$")
plt.show()
```
