```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
sns.set_style("whitegrid")
sns.set_theme(context='paper', palette="rocket")
```

```python
times=[10898, 22661]
cases = ["Caso 1", "Caso 2"]
df = pd.DataFrame({
    "case" : cases,
    "times" : times
})

ax = sns.barplot(df, x="case", y="times", estimator="sum")
plt.title("Tiempos de ejecución de cada caso de prueba")
plt.ylabel("Tiempo [s]")
plt.xlabel("")

for container in ax.containers:
    ax.bar_label(container)

plt.show()
```
