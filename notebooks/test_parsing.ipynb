```python
import argparse
```

```python
parser = argparse.ArgumentParser(
    prog="General Instance Parser"
)

parser.add_argument("--algorithm", choices=["sori", "itc"])

group_sori = parser.add_argument_group("SORI args")
group_itc = parser.add_argument_group("ITC arguments")

# SORI algorithm arguments
group_sori.add_argument("nm_students", type=int)
group_sori.add_argument("nm_courses", type=int)
group_sori.add_argument("nm_rooms", type=int)
group_sori.add_argument("nm_timeslots", type=int)
group_sori.add_argument("rho", type=int)
group_sori.add_argument("sigma_class", type=int)
group_sori.add_argument("output_path")

# ITC algorithm arguments
group_itc.add_argument("XML_instance")
group_itc.add_argument("model_dir")
group_itc.add_argument("output")
group_itc.add_argument("nm_timeslots", type=int)
group_itc.add_argument("room_cap", type=int)
group_itc.add_argument("class_lim", type=int)
```

```python
parser.parse_args()
```
